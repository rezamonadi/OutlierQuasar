```python
from astropy.table import Table, Column
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import pandas as pd
from scipy.io import loadmat   
df_0 = pd.read_csv('../data/data_0.csv')
df_scaled = pd.read_csv('../data/data_scaled.csv')
tab = Table.read('../data/reduced_dr16q.fits')
index = loadmat('../dbscan/dbscan-idx-eps-1.5-mpt-10.mat')
index=index['idx']
labels=[]
for i in range(len(index)):
    labels.append(index[i].item(0))
labels=np.array(labels)
FIBERID = tab['FIBERID']
MJD = tab['MJD']
PLATE = tab['PLATE']
Z = tab['Z']
```

```python
for cl in range(1,14):
    if(cl!=0):
        mask = (labels==cl)
        this_z = Z[mask]

        plt.hist(this_z, density=True, alpha=0.3, label='cl-%d'%cl)
plt.title('Redshift distribution for different clusters')
plt.xlabel('Redshidt')
plt.legend()
plt.savefig('redshift-dist.png', dpi=200)
plt.show()
```
